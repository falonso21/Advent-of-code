## Part 2

```python
def check_year(dictionary,field,min_,max_):
    if len(dictionary[field])==4:
        return min_<= int(dictionary[field]) <= max_
    else:
        return False
    
def check_height(dictionary):
    if ('cm' in dictionary['hgt']):
        return 150<= int(re.findall(r'\d+',dictionary['hgt'])[0]) <=193
    elif ('in' in dictionary['hgt']):
        return 59<= int(re.findall(r'\d+',dictionary['hgt'])[0]) <=76
    else:
        return False
    
def check_hair_color(dictionary):
    if ('#' in dictionary['hcl']) & (len(re.findall(r'\w+',dictionary['hcl'])[0])==6):
        return True
    else:
        return False

def check_eye_color(dictionary):
    return dictionary['ecl'] in ['amb','blu','brn','gry','grn','hzl','oth']

def check_passport(dictionary):
    try:
        return len(re.findall(r'\d+',dictionary['pid'])[0])==9
    except IndexError:
        return False
```

```python
cont = 0
for element in my_dict:
    if len(element)==8:
        if (check_year(element,'byr',1920,2002))&(check_year(element,'iyr',2010,2020))&\
        (check_year(element,'eyr',2020,2030))&(check_height(element))&(check_hair_color(element))\
        &(check_eye_color(element))&(check_passport(element)):
            cont+=1
    elif len(element)==7:
        try:
            element['cid']
        except KeyError:
            if (check_year(element,'byr',1920,2002))&(check_year(element,'iyr',2010,2020))&\
            (check_year(element,'eyr',2020,2030))&(check_height(element))&(check_hair_color(element))\
            &(check_eye_color(element))&(check_passport(element)):
                cont+=1
    else:
        pass
```
