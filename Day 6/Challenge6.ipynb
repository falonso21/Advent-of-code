## Day 6: Custom Customs

```python
import json
import re
```

```python
f = open("Challenge6.txt", "r")
data = f.read()
print(data)
```

```python
## We convert the raw text to a list of list with the groups, in each group at the same time each person is also a list
data = re.sub(r'(\w+)',r'"\1"',data)
data = '[[[' + data + ']]]'
data = data.replace('\n\n',']],[[')
data = data.replace('\n','],[')
my_list = json.loads(data)
my_list
```

## Part 1

```python
## In the following list of tuples we store tuples as follows: 
## (all answers of the group, len of unique elements)
## for the second item of the tuple just remember that set type drops duplicates so all we need to do is len the set
## of the first element
tuple_yes_answers = [(''.join(sum(x, [])),len(set((''.join(sum(x, [])))))) for x in my_list]
tuple_yes_answers
```

```python
sum(x[1] for x in tuple_yes_answers)
```

## Part 2

```python
## We add a element to the tuple: the number of member of the group
triplet_yes_answers = [(''.join(sum(x, [])),len(set((''.join(sum(x, []))))),len(x)) for x in my_list]
triplet_yes_answers
```

```python
## Checking all the memebers of the group have a 'yes answer' is equivalent to check
## if that question appears len(group) times in the concat solution so:
cont = 0
for x in triplet_yes_answers:
    for letter in list(set(x[0])):
        if x[0].count(letter)==x[2]:
            cont +=1 
cont 
```
